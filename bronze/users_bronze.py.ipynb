```python
entity = "users"
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
```

```python
df_users_bronze_raw = (
    spark.read.format('json').load(f'/Volumes/slack_etl/landing/raw/{entity}/')
    )
```

```python
df_users_bronze_enriched = df_users_bronze_raw.withColumns({
    'ingest_ts' : current_timestamp(),
    'source_file' :col("_metadata.file_path")
})
```

```python
(
    df_users_bronze_enriched.write.format('delta')
    .option("delta.enableChangeDataFeed", "true")
    .mode('overwrite')
    .saveAsTable(f'slack_etl.bronze.{entity}')
 )
```
